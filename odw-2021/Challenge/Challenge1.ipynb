```python
from gwpy.timeseries import TimeSeries
data = TimeSeries.read('challenge1.gwf', 'H1:CHALLENGE1')
```

```python
import numpy as np
signal=np.array(data)
print(data)
```

## The sampling rate of the data is given by dt = 0.000244140625 s or 1/dt = 4096 Hz

```python
#print(signal[-1])
#print(signal[0])
(len(signal)-1)*(0.000244140625)
```

## The duration of the data is ~64 secs

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(signal)
```

```python
specgram = data.spectrogram2(fftlength=4, overlap=2, window='hann') ** (1/2.)
plot = specgram.plot()
```

```python
ax = plot.gca()
ax.set_yscale('log')
ax.set_ylim(10, 1400)
ax.set_xlim(-1, 6)
ax.colorbar(
    clim=(1e-24, 1e-20),
    norm="log",
    label=r"Strain noise [$1/\sqrt{\mathrm{Hz}}$]",
)
plot  # refresh
```

## Plot of the data in time domain is represented above

```python
from gwosc.datasets import event_gps
from gwpy.timeseries import TimeSeries

data_Challenge1_H1_asd = data.asd(fftlength=4, method="median")
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(13,7))
# and plot using standard colours
ax.plot(data_Challenge1_H1_asd, label='Challenge1_H1', color='blue')

ax.set_xlim(2, 14000)
ax.set_ylim(5e-24, 2e-23)
# update the Livingston line to use standard colour, and have a label

ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylabel(r'Strain noise [$1/\sqrt{\mathrm{Hz}}$]')
ax.set_xlabel(r'Frequency [Hz]')

ax.legend()
plt.show()
```

```python
specgram_Challenge1_H1 = data.spectrogram2(fftlength=4, overlap=2, window='hann') ** (1/2.)

plot = specgram_Challenge1_H1.plot()

ax = plot.gca()
ax.set_yscale('log')
ax.set_ylim(10, 1400)
ax.colorbar(
    clim=(1e-24, 1e-20),
    norm="log",
    label=r"Strain noise [$1/\sqrt{\mathrm{Hz}}$]",
)
plot  # refresh
```

```python
hq = data.q_transform(frange=(30, 500), qrange=(80, 110))
plot = hq.plot()
ax = plot.gca()
#ax.set_epoch(gps)
ax.set_yscale('log')
ax.colorbar(label="Normalised energy")
```

```python
hq2 = data.q_transform(frange=(30, 500), qrange=(50, 110), outseg=(0,5)) 
plot = hq2.plot()
ax = plot.gca()
#ax.set_epoch(gps)
ax.set_yscale('log')
ax.colorbar(label="Normalised energy")
```

## Q-transform of the data identifying the signal

```python
hq2 = data.q_transform(frange=(30, 500), qrange=(5, 10), outseg=(0,5)) 
plot = hq2.plot()
ax = plot.gca()
#ax.set_epoch(gps)
ax.set_yscale('log')
ax.colorbar(label="Normalised energy")
```

# It is clear from the above two pictures that as we are dealing with a BBH merger using a qrange smaller than 10 produces a much sharper image than using a larger qrange(>50) which is more suitable for BNS mergers.

# The BBH merger takes place at ~2.95 secs

# An alternative approach to detecting signal using scipy

```python
import scipy

f, t, Sxx = scipy.signal.spectrogram(signal, 4096)
```

```python
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlim(left=0.5,right=3.5)
plt.ylim(bottom=0,top=500)
plt.xlabel('Time [sec]')
plt.show()
```

# Alternative approach using pycbc

```python
from pycbc.frame import read_frame

ts = read_frame("challenge1.gwf", "H1:CHALLENGE1")

print("Duration: {}s delta_t: {}s sampling rate: {}Hz Start: {} End: {}".format(ts.duration, ts.delta_t, 1/ts.delta_t,
                                              int(ts.start_time),
                                              int(ts.end_time)))
```

# The sampling rate is 4096 Hz, and the duration of the data is 128 secs

```python
import pylab

pylab.plot(ts.sample_times, ts)
pylab.ylabel('Strain_Challenge1')
pylab.xlabel('Time (s)')
pylab.show()
```

# Plot of the data in time domain

```python
t, f, p = ts.qtransform(.001, logfsteps=100, qrange=(8, 8), frange=(20, 1024))
pylab.figure(figsize=[10, 7])
pylab.title('q-transform of challenge_1 data')
pylab.pcolormesh(t, f, p**0.5, vmin=1, vmax=6)
pylab.yscale('log')
pylab.xlabel('Time (s)')
pylab.ylabel('Frequency (Hz)')
pylab.xlim(0 - 1, 0 + 5)
pylab.show()
```

# The merger occurs roughly at 2.95 or to round off at ~3 secs
